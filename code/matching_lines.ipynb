```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import os

import plot_utils as pu
import utils as ut
import data

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'monospace'
```

```python
name = 'starprob2_norm'
filename = f'{name}.csv'
fold_data = '../results/'
df = pd.read_csv(os.path.join(fold_data, filename))

size, rat = 720, 2
fig_w, fig_h = size*rat, size
subplots = (3,2)
fig, axs, fs, gs = pu.initialize_figure(fig_w=fig_w, fig_h=fig_h, ratio=None, 
                                        theme=None, subplots=subplots, text_size=0.75,
                                        hmerge=[0], hr=[1.5,1,1])
lw, ts, ss = 0.1, 1.2, 0.7

x = df['wavelength']*1e1
y = df['flux']


colors = {
    'H$': 'blue',       # Hydrogen
    'HeI': 'green',      # Helium
    'HeII': 'forestgreen', 
    'CaII': 'purple',     # Calcium
    'FeI': 'sienna',     # Iron
    'FeII': 'chocolate',     # Iron
    'MgII': 'crimson',       # Magnesium
    'NaI': 'darkred',     # Sodium
    'SiIII': 'steelblue',       # Silicon
    'SiIV': 'steelblue',       # Silicon
    'SiI': 'steelblue',       # Titanium
    'CII': 'orange',         # Carbon
    'OII': 'brown'        # Oxygen
}


axs[0][0].plot(x, y, color='k', lw=lw*fs)
spectral_lines = ut.sort_dic(data.spectral_lines_p2, lims=[None, None])
xl = [3700, 7250]
axs[0][0].set_xlim(xl[0], xl[1])
pu.plot_lines(axs[0][0], spectral_lines, colors, xl, fs)

axs[1][0].plot(x, y, color='k', lw=lw*fs)
spectral_lines = ut.sort_dic(data.spectral_lines_p2_balmer, lims=[None, None])
xl = [3700, 4190]
axs[1][0].set_xlim(xl[0], xl[1])
pu.plot_lines(axs[1][0], spectral_lines, colors, xl, fs)

axs[1][1].plot(x, y, color='k', lw=lw*fs)
spectral_lines = ut.sort_dic(data.spectral_lines_p2, lims=[None, None])
xl = [4250, 5100]
axs[1][1].set_xlim(xl[0], xl[1])
pu.plot_lines(axs[1][1], spectral_lines, colors, xl, fs)

axs[2][0].plot(x, y, color='k', lw=lw*fs)
spectral_lines = ut.sort_dic(data.spectral_lines_p2, lims=[None, None])
xl = [5820, 5950]
axs[2][0].set_xlim(xl[0], xl[1])
pu.plot_lines(axs[2][0], spectral_lines, colors, xl, fs)

axs[2][1].plot(x, y, color='k', lw=lw*fs)
spectral_lines = ut.sort_dic(data.spectral_lines_p2, lims=[None, None])
xl = [6500, 7120]
axs[2][1].set_xlim(xl[0], xl[1])
pu.plot_lines(axs[2][1], spectral_lines, colors, xl, fs)



axs[0][0].set_title(f'Problem Star II', fontsize=fs*ts*1.4)
flatten = lambda l: [item for sublist in l for item in sublist]
for ax_ in flatten(axs):
     ax_.tick_params(axis='both', which='major', pad=0.5*fs*ts,
                     labelsize=fs*ts)
axs[2][0].set_xlabel('Wavelength [$\AA$]', fontsize=fs*ts)
axs[2][1].set_xlabel('Wavelength [$\AA$]', fontsize=fs*ts)

savefold = '../figures/'
if not os.path.exists(savefold):
    # create the folder if it does not exist
    os.makedirs(savefold)
figname = f'{name}_lines.png'
savepath = savefold + figname

fig.savefig(savepath, dpi=300, bbox_inches='tight')
```
